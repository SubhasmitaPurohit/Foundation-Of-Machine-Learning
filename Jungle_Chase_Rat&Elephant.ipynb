## One_Hot Encoding

```python
onehotencoder=OneHotEncoder(categorical_features='all')
```

```python
encoded=onehotencoder.fit_transform(data).toarray()
```

```python
data_encoded=pd.DataFrame(encoded)
data_encoded.info()
```

```python
data_encoded['class']=df['class']
```

```python
data_encoded.info()
```

```python
data_encoded.head()
```

## Logestic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
x=data_encoded.columns[:-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(data_encoded[x],data_encoded['class'])
```

```python
print(x_train.shape,y_train.shape)
```

```python
print(x_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
predicted_class=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,predicted_class)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred=model.predict(x_test)
```

```python
confusionmatrix=confusion_matrix(y_test,y_pred)
confusionmatrix
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
neural_network_model = MLPClassifier(hidden_layer_sizes=(40,60))
```

```python
neural_network_model.fit(x_train,y_train)
```

```python
neural_network_predicted_class=neural_network_model.predict(x_test)
```

```python
accuracy_score(y_test,neural_network_predicted_class)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,neural_network_predicted_class))
```
